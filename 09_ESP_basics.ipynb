## Event streaming 

An __event__: the entity's observable state updates over time

### Common event formats

- __A primitive__ data format
such as a plain text, number, or date.
- A __key-value__ format - Python dict() -
and its value can be a primitive data type,
or complex data type like list,
tuple, JSON, XML, or even bytes.
For example, the GPS coordinates
of a car with car_id_1 as a tuple.
- A __key-value format associated with a timestamp__
to make it time-sensitive.
For example, the blood pressure of
a patient with ID pt001 as a tuple.
 

### Event streaming types

As te second diagram indicates, streaming can be really complicated with multiple
distributed event sources and destinations:
- data transfer pipelines may be
based on different communication protocols,
such as FTP, file transfer protocol,
HTTP, Hypertext Transfer Protocol,
JDBC, Java database connectivity,
SCP, secure copy, etc.
- an event destination can also be
an event source simultaneously.
For example, one application could receive
an event stream and process it,
then transport the processed results
as an event stream to another destination.

To overcome such a challenge of
handling different event sources and destinations,
we will need to employ the event streaming platform (ESP). 

## Event streaming platform (ESP) components

- __Event broker__
is designed to receive and consume events.
- __Event storage__ is used for storing
events being received from event sources.
Accordingly, event destinations do not need
to synchronize with event sources,
and stored events can be retrieved at will.
- __Analytic and query engine__,
which is used for querying and
analyzing the stored events.


__Event broker__ is the core component of an ESP
