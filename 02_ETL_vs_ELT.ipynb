# ELT process
_Based on the IBM Data Engineering Professional Certificate, ETL and Data Pipelines with Shell, Airflow and Kafka <br> Images' copyright: IBM Skills Network

ELT stands for __extract, load, and transform__.
- it is similar to
ETL in that similar stages are involved,
but the order in which they are performed is different.
For ELT processes, data is acquired and
directly loaded as is into its destination environment. 
-  From its destination environment ("a new home"), usually a sophisticated
analytics platform such as a data lake,
it can be transformed on demand. <br>

Like ETL, the first stage in
the ELT process is __extraction.__
- The extraction process obtains
the data from all sources and reads
the data often in
an asynchronous fashion into an application.<br>


The __loading__ process takes the raw data as is and loads it
into its new environment where
modern analytics tools can then be used directly.


The __transformation__ process for ELT is
much more dynamic than it is for conventional ETL. <br>

Modern analytics tools in
the destination environment enable
interactive, on-demand exploration
and visualization of data,
including advanced analytics,
such as modeling and prediction.

__Use cases__ for ELT processes typically
fall within the high performance computing
and big data realms

- dealing with the massive swings
and scale that come with implementing big data products,
- calculating real-time analytics on streaming big data,
- bringing together data sources
that are highly distributed around the globe

In terms of speed,
moving data is usually more
of a bottleneck than processing it.
The less you move it, the better.


__ELT is an emerging trend__

__Advantages: flexibility,
speed, and scalability__
- ELT is a flexible option that enables
a variety of applications from the same source of data.
- It is working with
a replica of the source data, so
there is __no information loss__.
Many kinds of transformations
can lead to information loss
and if these happen somewhere upstream in the pipeline,
it may be a long time before you
can have a change request met.
In the worst case, the information may be
forever lost if the raw data is not stored.
- __Cloud-based analytics platforms__ are ideally suited for
handling big data and
ELT processes in a cost-efficient manner,
and ELT is an emerging trend
mainly because cloud platform
technologies are enabling it. 

## Comparing ETL and ELT

__Favoring modern ELT over conventional ETL:__
- One of the factors driving that evolution is the demand to __release raw data__ to
a wider user base for the enterprise, along with the
ever-increasing ease of use and connection capabilities of analytics tools.
- Traditionally, ETL processes include an intermediate storage
facility called __a staging area__ -  a holding area for raw, extracted data where you can run processes prior to loading the resulting transformed data into a data warehouse or a data mart.

- The staging area fits the description of __a data lake__,
which is a modern self-serve repository for storing and manipulating raw data.
- A traditional staging area, however,
is not something that is usually shared across the company.
Its __a private siloed area__ set aside for developing, monitoring and
performance tuning
the data pipeline and its built-in transformations. 
- ELT work better with __Big Data__ and enable __a shorter time-to-insight__.
