# ELT process
_Based on the IBM Data Engineering Professional Certificate, ETL and Data Pipelines with Shell, Airflow and Kafka_ Images' copyright: IBM BM Skills Network

ELT stands for __extract, load, and transform__.
- it is similar to
ETL in that similar stages are involved,
but the order in which they are performed is different.
For ELT processes, data is acquired and
directly loaded as is into its destination environment. 
-  From its destination environment ("a new home"), usually a sophisticated
analytics platform such as a data lake,
it can be transformed on demand. <br>

Like ETL, the first stage in
the ELT process is __extraction.__
- The extraction process obtains
the data from all sources and reads
the data often in
an asynchronous fashion into an application.<br>


The __loading__ process takes the raw data as is and loads it
into its new environment where
modern analytics tools can then be used directly.


The __transformation__ process for ELT is
much more dynamic than it is for conventional ETL. <br>

Modern analytics tools in
the destination environment enable
interactive, on-demand exploration
and visualization of data,
including advanced analytics,
such as modeling and prediction.

__Use cases__ for ELT processes typically
fall within the high performance computing
and big data realms
